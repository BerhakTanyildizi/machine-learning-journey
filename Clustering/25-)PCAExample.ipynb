```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer(as_frame = True)
```

```python
df = data.frame
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns = 'target' , axis = 1 )
y = df['target']

X_train , X_test , y_train , y_test = train_test_split(X , y , random_state = 15 , test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler 

columns = X_train.columns

scaler = StandardScaler()

X_train = pd.DataFrame((scaler.fit_transform(X_train) ) , columns = columns)
X_test = pd.DataFrame((scaler.transform(X_test) )  , columns = columns)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import GradientBoostingClassifier
```

```python
Logistic = LogisticRegression()
Gradient = GradientBoostingClassifier()

Logistic.fit(X_train ,y_train)
Gradient.fit(X_train , y_train)

Logistic_y_pred = Logistic.predict(X_test)
Gradient_y_pred = Gradient.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report

print('LOGICTIC REGRESSION  ')
print('ACCURACY : '  , accuracy_score(y_test , Logistic_y_pred))
print('CLASSICICATION REPORT : \n' , classification_report(y_test , Logistic_y_pred))
print('\nCONFUSION MATRIX : \n' , confusion_matrix(y_test , Logistic_y_pred))

print('---------------------------\n\n')

print('GRADIENT BOOSTING CLASSIFIER  ')
print('ACCURACY : '  , accuracy_score(y_test , Gradient_y_pred))
print('CLASSICICATION REPORT : \n' , classification_report(y_test , Gradient_y_pred))
print('\nCONFUSION MATRIX : \n' , confusion_matrix(y_test , Gradient_y_pred))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 4 )

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

X_train_pca = pd.DataFrame(X_train_pca , columns = ['PC 1' , 'PC 2' , 'PC 3' , 'PC 4'])
X_test_pca = pd.DataFrame(X_test_pca , columns = ['PC 1' , 'PC 2' , 'PC 3' , 'PC 4'])

X_train_pca
```

```python
Logistic = LogisticRegression()
Gradient = GradientBoostingClassifier()

Logistic.fit(X_train_pca ,y_train)
Gradient.fit(X_train_pca , y_train)

Logistic_y_pred = Logistic.predict(X_test_pca)
Gradient_y_pred = Gradient.predict(X_test_pca)

print('LOGICTIC REGRESSION  ')
print('ACCURACY : '  , accuracy_score(y_test , Logistic_y_pred))
print('CLASSICICATION REPORT : \n' , classification_report(y_test , Logistic_y_pred))
print('\nCONFUSION MATRIX : \n' , confusion_matrix(y_test , Logistic_y_pred))

print('---------------------------\n\n')

print('GRADIENT BOOSTING CLASSIFIER  ')
print('ACCURACY : '  , accuracy_score(y_test , Gradient_y_pred))
print('CLASSICICATION REPORT : \n' , classification_report(y_test , Gradient_y_pred))
print('\nCONFUSION MATRIX : \n' , confusion_matrix(y_test , Gradient_y_pred))
```
