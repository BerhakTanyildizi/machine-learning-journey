```python
import seaborn as sns 
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/income-classification/income_evaluation.csv')
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
categorical = [col for col in df.columns if df[col].dtype == 'O']
numerical = [col for col in df.columns if df[col].dtype != 'O']
```

```python
categorical
```

```python
for col in df.columns : 
    print('\n',df[col].value_counts())
    
```

```python
for col in ['workclass' , 'occupation' , 'native-country'] :
    df[col] = df[col].replace(' ?' , np.nan)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns = 'income' , axis = 1 )
y = df['income']
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.3 , random_state = 15)
```

```python
X_train.head()
```

```python
for col in ['workclass' , 'native-country' , 'occupation'] : 
    X_train[col] = X_train[col].fillna(X_train[col].mode()[0])
    X_test[col] = X_test[col].fillna(X_train[col].mode()[0])
```

```python
X_test.isnull().sum()
```

```python
y_train = y_train.apply(lambda x : 1 if x.strip() == '>50K' else 0)
```

```python
y_test = y_test.apply(lambda x : 1 if x.strip() == '>50K' else 0)
```

```python
y_train.value_counts()
```

```python
target_means = y_train.groupby(X_train['native-country']).mean()

for i in[X_train , X_test] : 
    i['native-country-encoded'] = i['native-country'].map(target_means)
    i['native-country-encoded'] = i['native-country-encoded'].fillna(y_train.mean())
```

```python
X_train.head()
```

```python
for i in[X_train , X_test] : 
    i.drop(columns = 'native-country',axis = 1 , inplace = True )
```

```python
one_hot_enc = ['workclass' , 'marital-status' , 'relationship' , 'race' , 'sex' , 'occupation']

X_train_enc = pd.get_dummies(data = X_train , columns = one_hot_enc , drop_first = True , dtype = int)
X_test_enc = pd.get_dummies(data = X_test , columns = one_hot_enc  , drop_first = True , dtype = int)
```

```python
X_train_enc.head()
```

```python
# TRAIN 
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
```

```python
X_train_enc.drop(columns = 'education' , axis = 1 , inplace = True)
X_test_enc.drop(columns = 'education' , axis = 1 , inplace = True)

X_train = scaler.fit_transform(X_train_enc)
X_test = scaler.transform(X_test_enc)

X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train , y_train)
```

```python
y_pred = rfc.predict(X_test)

from sklearn.metrics import accuracy_score , confusion_matrix , classification_report 

print('Accuracy : \n' , accuracy_score(y_test , y_pred))
print('Classification Report :  \n ' , classification_report(y_test , y_pred))
print('Confusion Matrix : \n' , confusion_matrix(y_test , y_pred))
```

```python
# Hyperparamather Tuning : 
```

```python
from sklearn.model_selection import RandomizedSearchCV , StratifiedKFold
cv = StratifiedKFold(n_splits = 5)
params = {
    'n_estimators' : [100 , 300 , 600 , 1000],
    'max_depth' : [3 ,5 , 7 ,9 , 15 ,None],
    'criterion' : ['gini' , 'entropy' , 'log_loss'],
    'max_features' : ['sqrt' , 'log2' , None],
    'n_jobs' : [-1]
}
```

```python
random = RandomizedSearchCV(estimator = RandomForestClassifier(class_weight = 'balanced') , param_distributions = params , scoring = 'accuracy' , cv = cv , n_jobs = -1 )
random.fit(X_train , y_train)
y_pred = random.predict(X_test)

print('Accuracy : \n' , accuracy_score(y_test , y_pred))
print('Classification Report :  \n ' , classification_report(y_test , y_pred))
print('Confusion Matrix : \n' , confusion_matrix(y_test , y_pred))
```
