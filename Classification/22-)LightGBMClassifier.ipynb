```python
cols_to_drop = ['deck' , 'embark_town' , 'alive' , 'survived']
df['age'] = df['age'].fillna(df['age'].mean())
df['adult_male'] = df['adult_male'].astype(int)
df['alone'] = df['alone'].astype(int)
```

```python
df.info()
```

```python
df['class'].value_counts()
```

```python
X = df.drop(columns = cols_to_drop , axis = 1 )
y = df['survived']

from sklearn.model_selection import train_test_split

X_train , X_test , y_train ,y_test = train_test_split(X , y , test_size = 0.25 ,random_state = 15)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
X_train.info()
```

```python
one_hot_col = ['sex' , 'embarked' , 'class' , 'who']

transformer = ColumnTransformer(transformers = [
    ('onehot' , OneHotEncoder(drop = 'first' , handle_unknown = 'ignore' , dtype = int) , one_hot_col)
] , remainder = 'passthrough')
```

```python
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
```

```python
col_name = transformer.get_feature_names_out()

X_train = pd.DataFrame(X_train , columns = col_name)
X_test = pd.DataFrame(X_test , columns = col_name)
```

```python
X_test.head()
```

```python
import lightgbm as lgbm 
```

```python
lgbmc = lgbm.LGBMClassifier(verbosity = -1)
lgbmc.fit(X_train , y_train)
y_pred = lgbmc.predict(X_test)

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix 

print('Accuracy : ' , accuracy_score(y_test , y_pred))
print('Classification Report : \n' ,classification_report(y_test , y_pred))
print('Confusion Matrix : \n' , confusion_matrix(y_test , y_pred))
```

```python
lgbmc.feature_importances_
```

```python
importance = lgbmc.feature_importances_
feature_name = X_train.columns

importance_df = pd.DataFrame(
    {
        'Feature' : feature_name,
        'Importances' : importance
    }
).sort_values(by = 'Importances')
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {
    'max_depth' : [5 , 6 , 7 ,8 , 10 ], 
    'n_estimators' : [100 , 200 , 300],
    'num_leaves' : [32 , 64 ,128 , 256],
    'learning_rate' : [0.01 , 0.1 , 0.3 , 0.6],
    'min_data_in_leaf' : [2 , 4 ,8 ,10 ,14],
}
```

```python
random = RandomizedSearchCV(estimator = lgbm.LGBMClassifier(verbose = 0) , param_distributions = params , verbose = 0, cv = 5 , 
                            scoring = 'accuracy' , n_jobs = -1 )
```

```python
random.fit(X_train , y_train)
y_pred = random.predict(X_test)

print('Accuracy : ' , accuracy_score(y_test , y_pred))
print('Classification Report : \n' ,classification_report(y_test , y_pred))
print('Confusion Matrix : \n' , confusion_matrix(y_test , y_pred))
```
