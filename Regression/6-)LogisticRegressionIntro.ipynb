```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('6-bank_customers.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Datamız temiz data temziliğine gerek yok hatta değerler zaten scale edilmiş olduğu için scale de etmeyceğiz
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('subscribed' , axis = 1)
y = df['subscribed']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , random_state = 15 , test_size = 0.3)
```

```python
# Scale yapmıyoruz veriler zaten scale edilmiş halde
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(X_train , y_train)
y_pred = logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
print('Accuary Score : ' , accuracy_score(y_test , y_pred))
print('Classification Report :\n ' , classification_report(y_test , y_pred))
print('Confusion Matrix : \n' , confusion_matrix(y_test , y_pred)) 
```

```python
# Doğruluk oranı %92 
# Precision ve recall gayet yakın ve düzgün 
# F1 Score da yüksek 
# True Positive = 134  --- True Negative = 142 -->> Gayet başarılı tahminler ve istatistikler 
```

```python
# Hyperparameter Tuning 
```

```python
model = LogisticRegression()
```

```python
penalty = ['l1' , 'l2' , 'elasticnet']
c_values = [100 , 10 , 0 , 0.1 , 0.01]
solver = ['lbfgs' , 'liblinear' , 'newton-cg' , 'newton-cholesky' , 'sag' , 'saga']
```

```python
params  = dict(penalty = penalty , C = c_values , solver = solver)
```

```python
from sklearn.model_selection import GridSearchCV , StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator = model , param_grid = params , scoring = 'accuracy' , n_jobs=-1 , cv = 5 )
```

```python
grid
```

```python
grid.fit(X_train , y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
print('Accuary Score : ' , accuracy_score(y_test , y_pred))
print('Classification Report : \n' , classification_report(y_test , y_pred))
print('Confusion Matrix : \n' , confusion_matrix(y_test , y_pred)) 
```
