```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
X = df.drop(columns = 'selling_price' , axis = 1 )
y = df['selling_price']

X_train , X_test , y_train ,y_test = train_test_split(X , y , test_size = 0.3 , random_state = 15)
```

```python
df.head()
```

```python
for col in df.columns : 
   print(df[col].nunique())
```

```python
cols_to_freq = ['car_name' , 'brand' , 'model']
```

```python
X_train['car_name'].value_counts() / len(X_train)
```

```python
for col in cols_to_freq : 
    freq = X_train[col].value_counts() / len(X_train)
    X_train[col + '_freq'] = X_train[col].map(freq)
    X_test[col + '_freq'] = X_test[col].map(freq)

    mean_freq = freq.mean()
    X_test[col + '_freq'] = X_test[col + '_freq'].fillna(mean_freq) 
```

```python
X_train.drop(columns = cols_to_freq , axis = 1 , inplace = True)
X_test.drop(columns = cols_to_freq , axis = 1 , inplace = True)
X_test.head()
```

```python
# One Hot Encoding 
```

```python
cols_one_hot = ['seller_type' ,'fuel_type' ,'transmission_type']
```

```python
transformer = ColumnTransformer(
    transformers = [
        ('onehot' , OneHotEncoder(drop = 'first' , dtype = int , handle_unknown = 'ignore') , cols_one_hot)
    ], remainder = 'passthrough')
```

```python
X_train = pd.DataFrame(transformer.fit_transform(X_train) , columns = transformer.get_feature_names_out())
X_test = pd.DataFrame(transformer.transform(X_test) , columns = transformer.get_feature_names_out())
```

```python
X_train.info()
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor

ada = AdaBoostRegressor()
ada.fit(X_train  , y_train)
y_pred = ada.predict(X_test)

from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error

print('R2 : ' , r2_score(y_test , y_pred ))
print('MSE : ' , mean_squared_error(y_test , y_pred))
print('MAE : ' , mean_absolute_error(y_test , y_pred))
```

```python
# Hyperparamether Tuning : 
from sklearn.model_selection import GridSearchCV 
```

```python
dtr = DecisionTreeRegressor()
params = {
    'estimator__max_depth' : [5 , 10 , 25 , 50 , 100 , None],
    'n_estimators' : [10 , 50 , 100 , 150 ], 
    'learning_rate' : [0.5 , 1.0 , 1.5 , 2.0],
    'loss' : ['linear' , 'square' , 'exponential']
}
grid = GridSearchCV(estimator = AdaBoostRegressor(DecisionTreeRegressor()) , param_grid = params , scoring = 'r2' , cv = 5 ,n_jobs = -1 )
```

```python
grid.fit(X_train, y_train)
y_pred = grid.predict(X_test)

print('R2 : ' , r2_score(y_test , y_pred ))
print('MSE : ' , mean_squared_error(y_test , y_pred))
print('MAE : ' , mean_absolute_error(y_test , y_pred))
```
