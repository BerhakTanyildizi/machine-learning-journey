```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv('7-cyber_attack_data.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
X = df.drop('attack_type' , axis = 1) 
y = df['attack_type']

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , random_state = 15 , test_size = 0.3 )

from sklearn.linear_model import LogisticRegression 

logistic = LogisticRegression()
logistic.fit(X_train , y_train)

y_pred = logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
score = accuracy_score(y_test , y_pred) 
classification = classification_report(y_test , y_pred)
matrix = confusion_matrix(y_test , y_pred)
```

```python
print('Score : ' , score)
print('Classification Report : \n' , classification)
print('Matrix : \n' , matrix)
```

```python
# Hypermarameter Tuning Yapalım : 
```

```python
model = LogisticRegression()
```

```python
penalty = ['l1' , 'l2' , 'elasticnet']
c_values = [100 , 10 , 0 , 0.1 , 0.01]
solver =  ['lbfgs' , 'liblinear' , 'newton-cg' , 'newton-cholesky' , 'sag' , 'saga']
params = dict(penalty = penalty , C = c_values , solver = solver)
```

```python
from sklearn.model_selection import GridSearchCV , StratifiedKFold
```

```python
cv = StratifiedKFold(n_splits = 5 )
```

```python
grid = GridSearchCV(estimator = model , param_grid = params , cv = cv , n_jobs = -1)
```

```python
grid.fit(X_train , y_train)
```

```python
y_pred = grid.predict(X_test)

score = accuracy_score(y_test , y_pred) 
classification = classification_report(y_test , y_pred)
matrix = confusion_matrix(y_test , y_pred)

print('Score : ' , score)
print('Classification Report : \n' , classification)
print('Matrix : \n' , matrix)

# Sadece bazı skorlar biraz iyileşti ve conf matrix bir tık değişti ancak yeterli olmadı
```

```python
# One vs One  ----- One vs Rest deneyelim
```

```python
from  sklearn.multiclass import OneVsOneClassifier , OneVsRestClassifier
```

```python
model_for_multiclass = LogisticRegression()
```

```python
onevsoneclass = OneVsOneClassifier(estimator = model_for_multiclass , n_jobs = -1)
onevsrestclass = OneVsRestClassifier(estimator = model_for_multiclass , n_jobs = -1)
```

```python
# One vs One

onevsoneclass.fit(X_train , y_train)
y_pred = onevsoneclass.predict(X_test)

score = accuracy_score(y_test , y_pred)
matrix = confusion_matrix(y_test , y_pred)
classification = classification_report(y_test , y_pred)

print('Score : ' , score)
print('Classification Report : \n' , classification)
print('Matrix : \n' , matrix)
```

```python
# One vs Rest

onevsrestclass.fit(X_train , y_train)
y_pred = onevsrestclass.predict(X_test)

score = accuracy_score(y_test , y_pred)
matrix = confusion_matrix(y_test , y_pred)
classification = classification_report(y_test , y_pred)

print('Score : ' , score)
print('Classification Report : \n' , classification)
print('Matrix : \n' , matrix)
```
